```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/content/1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap = 'gray')
plt.show()
```

```python
img = cv2.imread('/content/1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

original_img = img
brightness_img = img.astype(np.float32) + 50
brightness_img = np.clip(brightness_img, 0, 255)
brightness_img = brightness_img.astype(np.uint8)

darkness_img = img.astype(np.float32) - 80
darkness_img = np.clip(darkness_img, 0, 255)
darkness_img = darkness_img.astype(np.uint8)

print('Original Image: ')
plt.imshow(original_img)
plt.show()
print('Brightness Image: ')
plt.imshow(brightness_img)
plt.show()
print('Darkness Image: ')
plt.imshow(darkness_img)
plt.show()
```
